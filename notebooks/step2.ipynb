# Typosquatting

```python
# Generate list of typosquatting domains
typosquatting = step2[step2["squatting_technique"] == "typosquatting"]
```

```python
# Number of typosquatting domains
len(typosquatting)
```

```python
# Get different typosquatting techniques

# function returns number of different chars between two strings
def count_char_diff(s1, s2):
    counter = 0
    for x, y in zip(s1, s2):
        if x != y:
            counter += 1
            
    return counter

typosquatting_types = {"prepend_www": 0, "omit_character": 0, "swap_characters": 0,\
                       "replace_qwerty": 0, "duplicate_character": 0}
for index, row in typosquatting.iterrows():
    if row["created_domain"].startswith("www"):
        typosquatting_types["prepend_www"] += 1
    elif len(row["created_domain"]) < len(row["reference_domain"]):
        typosquatting_types["omit_character"] += 1
    elif len(row["created_domain"]) > len(row["reference_domain"]):
        typosquatting_types["duplicate_character"] += 1
    elif count_char_diff(row["created_domain"], row["reference_domain"]) == 1:
        typosquatting_types["replace_qwerty"] += 1
    elif count_char_diff(row["created_domain"], row["reference_domain"]) == 2:
        typosquatting_types["swap_characters"] += 1
    else:
        print(row["created_domain"])
        print(row["reference_domain"])
typosquatting_types
```

```python
# Position of first changed character

# function returns the position of the first character which is different among the two strings
def get_first_diff_char_pos(s1, s2):
    s1 = s1.split(".")[0]
    s2 = s2.split(".")[0]
    pos = 0
    for x, y in zip(s1, s2):
        if x == y:
            pos += 1
        else:
            return pos
        
    return pos

positions = []
for index, row in typosquatting.iterrows():
    if not row["created_domain"].startswith("www"):
        # ignore prepend_www cases because the first character is already different
        pos = get_first_diff_char_pos(row["created_domain"], row["reference_domain"])
        # normalize position by dividing it by the domain length, ignoring the top-level domain
        positions.append(pos / len(row["created_domain"].split(".")[0]))
pd.Series(positions).describe()
```

# Homograph

```python
# Generate list of homograph domains
homograph = step2[step2["squatting_technique"] == "homograph"]
```

```python
# Number of homograph domains
len(homograph)
```

```python
# Which homograph pairs were used most often?
homograph_pairs = []
for index, row in homograph.iterrows():
    for x, y in zip(row["created_domain"], row["reference_domain"]):
        if x != y:
            homograph_pairs.append((x, y))
print("Most common homograph pairs:")
print(Counter(homograph_pairs).most_common()[:10])
print("Most often replaced characters:")
print(Counter([elem[1] for elem in homograph_pairs]).most_common()[:10])
```

```python
# Position of first changed character
positions = []
for index, row in homograph.iterrows():
    pos = get_first_diff_char_pos(row["created_domain"], row["reference_domain"])
    # normalize position by dividing it by the domain length, ignoring the top-level domain
    positions.append(pos / len(row["created_domain"].split(".")[0]))
pd.Series(positions).describe()
```

# Combosquatting

```python
# Generate list of combosquatting domains
combosquatting = step2[step2["squatting_technique"] == "combosquatting"]
```

```python
# Number of combosquatting domains
len(combosquatting)
```

```python
# Get combosquatting types (characters/terms in front/back)
terms = ["login", "secure", "account", "verify", "service", "official", "support", "validation", "authentication",\
         "mobile", "signin"]
combosquatting_types = {"chars_front_none_back": 0, "terms_front_none_back": 0,\
                        "none_front_chars_back": 0, "none_front_terms_back": 0,\
                        "chars_front_chars_back": 0, "terms_front_terms_back": 0,\
                        "chars_front_terms_back": 0, "terms_front_chars_back": 0}
for index, row in combosquatting.iterrows():
    key = ""
    if any([row["created_domain"].startswith(term + "-") for term in terms]):
        key = "terms_front_"
    elif row["created_domain"].startswith(row["reference_domain"].split(".")[0]):
        key = "none_front_"
    else:
        key = "chars_front_"
    if any ([row["created_domain"].split(".")[0].endswith("-" + term) for term in terms]):
        key +="terms_back"
    elif row["created_domain"].endswith(row["reference_domain"]):
        key += "none_back"
    else:
        key += "chars_back"
    combosquatting_types[key] += 1
combosquatting_types
```
